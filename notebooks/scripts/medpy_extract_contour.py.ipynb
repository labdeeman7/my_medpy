```python
from medpy.io import load
import matplotlib.pyplot as plt
import matplotlib.cm as cm

%matplotlib inline
```

Script to extract the contours from binary objects. Consider the following image:

```python
i, _ = load("resources/brainmask.nii.gz")
plt.imshow(i, cmap = cm.Greys_r)
```

To extract the 1-pixel wide contour, we simply run

```python
!medpy_extract_contour.py resources/brainmask.nii.gz output/contour.nii.gz -f
```

Which results in:

```python
o, _ = load("output/contour.nii.gz")
plt.imshow(o, cmap = cm.Greys_r)
```

This tool is, as all *medpy* scripts, dimensionality free. For more than two dimensions shells rather than contours are created.
