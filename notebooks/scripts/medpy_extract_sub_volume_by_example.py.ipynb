```python
from medpy.io import load
import matplotlib.pyplot as plt
import matplotlib.cm as cm

%matplotlib inline
```

Nifty script to crop an image by supplying a binary mask. Let's assume we have a brain scan and an associated binary mask like:

```python
i1, _ = load("resources/flair.nii.gz")
i2, _ = load("resources/brainmask.nii.gz")

f, axarr = plt.subplots(1, 2)
axarr[0].imshow(i1, cmap = cm.Greys_r)
axarr[1].imshow(i2, cmap = cm.Greys_r)
```

We can crop the original image using

```python
!medpy_extract_sub_volume_by_example.py resources/flair.nii.gz output/cropped.nii.gz resources/brainmask.nii.gz -f
```
